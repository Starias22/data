### Plot functions

#### cosinus

```python
import numpy as np
#cosinus function

x=np.linspace( -np.pi, np.pi,num=100)
y=np.cos(x)
plt.plot(x,y)
plt.show()
```

#### PDF of the standard normal distribution
$f(t)=\frac{1}{\sqrt{2\pi}}e^{-\frac{1}{2}t^2}$

```python

def f(x):
    return (1/np.sqrt(2*np.pi))*np.exp(-np.power(x,2))
x=np.linspace( -1, 1,num=100)
plt.plot(x,f(x))
plt.show()    
```

#### Function g $g(x)=-x+ln(x)$

```python
g= lambda x: -x+np.log(x)
x=np.linspace( 0, 5,num=10000)[1:]
plt.plot(x,g(x))
plt.show() 
```

Be careful when x is not a numpy array 

```python
h=lambda x: np.power(x,4)-x**2+2*x+3
x=range(-1,1000)
plt.plot(x,[h(i) for i in x])
plt.show() 
```

#### PDF of the uniform distribution with a=0 and b=1
 $\begin{align*}
f(x) = \begin{cases}
            1 & \text{if } 0 \le x \le 1\\
            0 & \text{else }
        \end{cases}
\end{align*}$

```python
f=lambda x:np.where(0<=x and x<=1,1,0) 
x=np.linspace(-10,10).tolist()
plt.plot(x,[f(i) for i in x])
plt.show() 
```

#### Function F: $\begin{align*}
F(x) = \begin{cases}
            0 & \text{if } x \lt 0\\
            \frac{x}{200} & \text{if } 0 \lt x \lt 100\\
            1 & \text{if } x \geq 100\\
        \end{cases}
\end{align*}$

```python
F=lambda x:np.where(0<x and x<100,x/100,np.where(x<0 ,0,1)) 
x=np.linspace(-1000,1000).tolist()
plt.plot(x,[F(i) for i in x])
plt.show() 
```

### Markers


```python
ycoords=[4,5,8,-20,5.55,4,-7.55]
#Let us default x from 0 to len(ycoords)-1
```

#### marker(the type  or shape of marker)


```python
#star
plt.plot(ycoords,marker='*',label='with star marker')
plt.legend()
plt.show()

#plus
plt.plot(ycoords,marker='P',color='m',label='with plus marker')#magenta
plt.legend()
plt.show()
```

#### marker properties
The following properties can be applied to markers

##### markersize or ms

```python
#ms or markersize

plt.plot(ycoords,marker='D')#marker in diamond with normal size
plt.show()

plt.plot(ycoords,marker='D',ms=30,color='y')#marker in diamond with 30 as size
plt.show()
```

##### markeredgecolor or mec

```python
##mec or markeredgecolor

plt.plot(ycoords,markeredgecolor='b',ms=30,marker='o',color='black')#blue as edge color
plt.show()
```

##### markerfacecolor or mfc##mfc or markerfacecolor

```python
plt.plot(ycoords,markerfacecolor='#4CAF50',ms=30,mec='m',marker='P')
plt.show()

plt.plot(ycoords,mfc='hotpink',ms=30,mec='m',marker='P')
plt.show()
```

## Grid

```python
y=[4,45,2,5,6,7,8,10]

plt.plot(y,marker='o')
plt.title('An example')
plt.xlabel('x-coordinates')
plt.ylabel('ycoordinates')
plt.grid()
plt.show()
```

### Grid axis:x, y, both

```python
##x axis
plt.plot(y,marker='o')
plt.title('An example')
plt.xlabel('x-coordinates')
plt.ylabel('ycoordinates')
plt.grid(axis='x')
plt.show()
```

```python
##y axis
plt.plot(y,marker='o')
plt.title('An example')
plt.xlabel('x-coordinates')
plt.ylabel('ycoordinates')
plt.grid(axis='y')
plt.show()
```

```python
##both(default)
plt.plot(y,marker='o')
plt.title('An example')
plt.xlabel('x-coordinates')
plt.ylabel('ycoordinates')
plt.grid(axis='both')
plt.show()
```

### Grid line properties: color, lw(for the weight of the line), linestyle(or ls)

```python
plt.plot(y,marker='X')
plt.title('An example')
plt.xlabel('x-coordinates')
plt.ylabel('ycoordinates')
plt.grid(axis='both',linestyle='--',lw=2.5,color='hotpink')
plt.show()
```


## Subplot



```python
x1=[4,5.33,10,11,13]
y1=[-2,5,3,6,-4.4]
x2=[4,5,6]
y2=[2,3,6]
plt.subplot(1,2,1)
plt.plot(x1,y1,color='hotpink',label='subplot 1',marker='*')
plt.title('Subplot1',loc='left',color='blue')
plt.legend(loc='lower left')

plt.subplot(1,2,2)
plt.plot(x2,y2,color='magenta',label='subplot 2',marker='P')
plt.title('Subplot2',loc='right',color='red')
plt.legend(loc='best')

plt.suptitle('Subplots example')

plt.show()
```
